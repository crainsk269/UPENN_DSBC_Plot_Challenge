### Analysis

1. Initially, we looked data which showed results from 10 different drugs on a combination of both male and female mice which varied in weight. Of these drugs tested, the average mouse weight was 27g, while the average weight for the other two drugs was 19g (Capomulin and Ramicane). Capomulin and Ramicane both finished in the top two amongst all other drugs for Tumor Volume (mm3) mean, median, var, std, and sem; showing that they were more successful than the remaining eight drugs.

2. Of the original 249 test subjects, mouse g989 had duplicate values within the Timepoint and was therefore excluded from further analysis in our studies.

3. Male and female populations were almost equal; 51% and 49% respectively. Additional analysis on male only and female only results was not conducted during our research.

4. Four drugs were selected for outlier analysis: Capomulin, Ramicane, Infubinol, and Ceftamin. Infubinol was the only one of the four which had outliers, it also had the highest numbers in all five categories of mean, median, var, std, and sem.

5. From all of the analysis that we conducted, Capomulin and Ramicane showed to be the two most succesful drugs at treating tumors in mice.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_df = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
mouse_study_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mouse_mean = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
mouse_median = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
mouse_var = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
mouse_std = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
mouse_sem = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
drug_regimen_stats = pd.DataFrame({'Mean Tumor Volume': mouse_mean,
                                 'Median Tumor Volume': mouse_median,
                                 'Tumor Volume Variance': mouse_var,
                                 'Tumor Volume Std. Dev.': mouse_std,
                                 'Tumor Volume Std. Err.': mouse_sem})


# Minor data wrangling
drug_regimen_stats.index.name = 'Drug Regimen'

# Display the DataFrame
drug_regimen_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
mouse_study_aggregate = mouse_study_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
mouse_study_aggregate
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = mouse_study_clean.groupby(['Mouse ID'])['Timepoint'].max()
last_timepoint_df = pd.DataFrame(last_timepoint)
# last_timepoint_df

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_timepoint_merged = pd.merge(last_timepoint_df, mouse_study_clean, on=['Mouse ID', 'Timepoint'])
last_timepoint_merged
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    # Locate the rows which contain mice on each drug and get the tumor volumes
for drug in drugs:
    tumor_df = last_timepoint_merged.loc[last_timepoint_merged['Drug Regimen'] == drug]['Tumor Volume (mm3)']
    
    tumor_volumes.append(tumor_df)
    
    quartiles = tumor_df.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outliers = tumor_df.loc[(tumor_df >= upper_bound) | (tumor_df <= lower_bound)]
    
    print(f"{drug}'s potential outliers: ", outliers)
    
    print('---------------------------------------------')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
tumor_distribution = dict(marker='o', markerfacecolor='r', markersize=12, markeredgecolor='black')

# Box plot
plt.boxplot(tumor_volumes, flierprops=tumor_distribution)

# Title and labels
plt.title('Tumor Volumes by Drug Types')
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks([1, 2, 3, 4], drugs)

# Display
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
mouse_weight = mouse_study_clean.groupby(['Mouse ID', 'Drug Regimen']).agg({'Tumor Volume (mm3)': 'mean', 'Weight (g)': 'mean'}).reset_index()
cap_reg = mouse_weight.loc[mouse_weight['Drug Regimen'] == 'Capomulin']

# Scatter plot
plt.scatter(cap_reg['Weight (g)'], cap_reg['Tumor Volume (mm3)'], marker='o')

#Title and labels
plt.title('Capomulin (Mouse Weight vs Average Tumor Volume)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# Correlation coefficient
corr_coef = cap_reg['Tumor Volume (mm3)'].corr(cap_reg['Weight (g)'])
print(f'The correlation between mouse weight and average tumor volume is {corr_coef:.2f}')

# Linear regression
slope, intercept, rvalue, pvalue, stderr = linregress(cap_reg['Weight (g)'], cap_reg['Tumor Volume (mm3)'])
regress_values = slope * cap_reg['Weight (g)'] + intercept
plt.plot(cap_reg['Weight (g)'], regress_values, 'r')

# Scatter plot
plt.scatter(cap_reg['Weight (g)'], cap_reg['Tumor Volume (mm3)'], marker='o')

# Title and labels
plt.title('Mouse Weight and Average Tumor Volume')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display
plt.show()
```
